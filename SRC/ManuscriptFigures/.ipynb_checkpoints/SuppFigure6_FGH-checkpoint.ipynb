```r
source("Main.R")
source("Conf.R")
source("Utilities.R")
library(stringr) 
```

```r
##Each group and double KO pair are treated as a separate covariate
combos <- readRDS("../Notebooks/CombinatorialPerturbations/RDSFiles/ComboEffects_lm_residuals.rds")

combos <- combos[combos$term %ni% c("(Intercept)", "n_genes", "mt_frac", paste0("leiden", 1:9)),]
combos = data.table(combos)
combos[,FDR:=p.adjust(p.value, method = "fdr", n = length(p.value)),by=respGene]
combos = data.frame(combos)
combosFDR = copy(combos)


combos[,c("std.error", "statistic", "p.value","FDR")] = NULL
combos = reshape(combos, idvar = "respGene", timevar = "term",  direction = "wide")
rownames(combos) = combos$respGene
combos$respGene = NULL
colnames(combos) = sapply( colnames(combos), function(x){strsplit(x,"estimate.")[[1]][2]})


combosFDR[,c("std.error", "statistic", "estimate","p.value")] = NULL
combosFDR = reshape(combosFDR, idvar = "respGene", timevar = "term",  direction = "wide")
rownames(combosFDR) = combosFDR$respGene
combosFDR$respGene = NULL
colnames(combosFDR) = sapply( colnames(combosFDR), function(x){strsplit(x,"FDR.")[[1]][2]})

combosFDR$respGene = rownames(combosFDR)
combosFDR = melt(combosFDR, id.vars="respGene")
colnames(combosFDR) = c("Gene", "Comb", "FDRObserved")
```

```r
head(combos)
```

```r
allPred = data.frame()

for( i in c("K_0", "K_1", "K_2", "K_3", "K_4", "K_5")){
      for ( j in c( "K_1", "K_2", "K_3", "K_4", "K_5")){
      
          if(paste0(i,"W",j) %in% colnames(combos)){
            prediction = combos[,i] + combos[,j]
            observed = combos[,paste0(i,"W",j)]
            myR2 = round(computeR2(response = observed, prediction = prediction), digits=2)
            myMAE = round(mean(abs(prediction - observed)), digits=3)
              
            myTemp =  data.frame("Predictedvalues" = prediction ,
                                 "Observedvalues" = observed,
                                 "Comb"=paste0(i,"W",j),
                                 "Gene" = rownames(combos),
                                 "R2" = myR2,
                                 "MAE" = myMAE,
                                 "corS" = round(cor(combos[,paste0(i,"W",j)], (combos[,i] + combos[,j]), method="spearman"), digits=2),
                                 "pearsonS" = round(cor(combos[,paste0(i,"W",j)], (combos[,i] + combos[,j]), method="pearson"), digits=2),
                                 "Group1_val" = combos[,i],
                                 "Group2_val" = combos[,j]
                                )
            allPred = rbind(allPred, myTemp)
          }
      }
}

allPred$Group1 = sapply(allPred$Comb, function(x){return(strsplit(as.character(x),"W")[[1]][1]) })
allPred$Group2 = sapply(allPred$Comb, function(x){return(strsplit(as.character(x),"W")[[1]][2]) })
allPred$ObsPredDif = allPred$Observedvalues - allPred$Predictedvalues
allPred$AbsObsPredDif = abs(allPred$Observedvalues - allPred$Predictedvalues)
allPred$GroupGene = paste0(allPred$Comb,"_",allPred$Gene)
```

```r
options(repr.plot.width=12, repr.plot.height=12)

allPred$Group1 <- sapply(allPred$Group1 , function(x){str_replace_all(as.character(x),"_","")})
allPred$Group2 <- sapply(allPred$Group2, function(x){str_replace_all(as.character(x),"_","")})

allPred[allPred$R2 < 0, "R2"] = 0

pdf("Figure_S6F.pdf", width=12, height=12)
ggplot(allPred, aes(x=Predictedvalues, y=Observedvalues)) + 
        geom_point(alpha=0.4, shape=20, color="blue") + 
        facet_grid(Group1 ~ Group2)+
        geom_text(aes(label= paste0("R^2 = ",R2), color="red"), x=0, y=-0.3, size=5)+
        geom_text(aes(label= paste0("MAE = ",MAE), color="red"), x=0, y=-0.5, size=5)+
        geom_abline(linetype="dashed")+
        theme_minimal()+
        theme(axis.text = element_text(size = 15),
              axis.title = element_text(size = 25),
              strip.text = element_text(size = 25), legend.position='none')+
        xlim(-0.6,0.6)+ylim(-0.6,0.6)+
        ggtitle("Explained variance by additive effects")+
        ylab("Observed fold change wrt control (lg2)")+
        xlab("Predicted fold change wrt control (lg2)")

dev.off()
```

```r
combosFDR = combosFDR[combosFDR$Comb %in% unique(allPred$Comb),]
allPred = merge(allPred, combosFDR, by=c("Gene", "Comb"))
allPred$interactionterm = paste0(allPred$Group1,":",allPred$Group2)
dim(allPred)
```

```r
combosInterP <- readRDS("./../Notebooks/CombinatorialPerturbations/RDSFiles/ComboEffects_lm_residuals_withInteractions.rds")
combosInterP <- combosInterP[combosInterP$term %ni% c("(Intercept)", "n_genes", "mt_frac", paste0("leiden", 1:9)),]
combosInterP = data.table(combosInterP)
combosInterP[,FDR:=p.adjust(p.value, method = "fdr", n = length(p.value)),by=respGene]
combosInterP = data.frame(combosInterP)
combosInterP$term <- sapply(combosInterP$term, function(x){str_replace_all(as.character(x),"_","")})

combosInterP <- combosInterP[,c("term", "estimate","FDR", "respGene")]
combosInterP = combosInterP[combosInterP$term %in% allPred$interactionterm,]
colnames(combosInterP) = c("interactionterm", "interactionEstimate","FDRInteraction", "Gene")

allPred = merge(allPred, combosInterP, by=c("interactionterm", "Gene"))
dim(allPred)
```
