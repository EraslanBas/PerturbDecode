```python
fig, axs = plt.subplots(2, 3)
fig.set_figheight(15)
fig.set_figwidth(15)

axs[0, 0].plot(errorDF.Epoch, errorDF.TrainingError, color='blue', label='Training error')
axs[0, 0].plot(errorDF.Epoch, errorDF.TestError, color='g', label='Test error')
axs[0, 0].set_title('Total error')
axs[0, 0].legend(loc="upper right")

axs[0, 1].plot(errorDF.Epoch, errorDF.Training_BCE_error, color='blue', label='Training BCE error')
axs[0, 1].plot(errorDF.Epoch, errorDF.Test_BCE_error, color='g', label='Test BCE error')
axs[0, 1].set_title('BCE error')
axs[0, 1].legend(loc="upper right")

axs[1, 0].plot(errorDF.Epoch, errorDF.Training_KLD_error, color='blue', label='Training KLD error')
axs[1, 0].plot(errorDF.Epoch, errorDF.Test_KLD_error, color='g', label='Test KLD error')
axs[1, 0].set_title('KLD error')
axs[1, 0].legend(loc="upper right")

axs[0, 2].plot(errorDF.Epoch, errorDF.Training_L1_error, color='blue', label='Training L1 error')
axs[0, 2].plot(errorDF.Epoch, errorDF.Test_L1_error, color='g', label='Test L1 error')
axs[0, 2].set_title('L1 error')
axs[0, 2].legend(loc="upper right")

# axs[1, 1].plot(errorDF.Epoch, errorDF.Training_class_error, color='blue', label='Training class error')
# axs[1, 1].plot(errorDF.Epoch, errorDF.Test_class_error, color='g', label='Test class error')
# axs[1, 1].set_ylim([0, 2000])
# axs[1, 1].set_title('Classification error')
# axs[1, 1].legend(loc="upper right")



plt.xlabel("Number of Epochs")

plt.legend();
```
