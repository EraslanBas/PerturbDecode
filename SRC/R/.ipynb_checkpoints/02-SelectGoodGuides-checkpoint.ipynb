```r
#myDistFlat <- readRDS("RDSFiles/betaCovPearsonFlat.rds")
myDistFlat <- readRDS("RDSFiles/GuideSelect_betaCovPearsonFlat.rds") 
```

```r
myDistFlat$sameGene <- factor(myDistFlat$sameGene, levels=c("TRUE", "FALSE"))

cList = list(c("TRUE", "FALSE"))

n_fun <- function(x){
  return(data.frame(y = median(x), label = paste0("n = ",median(x))))
}

options(repr.plot.width=8, repr.plot.height=6)
ggplot(myDistFlat, aes(x=sameGene, y=sDist)) + 
     geom_violin(trim=FALSE, aes(fill=sameGene))+
     geom_boxplot()+theme_bw()+ 
     stat_summary(fun.data = n_fun, geom = "text", size=6)+
        stat_compare_means(method="t.test",paired = FALSE,
                                comparisons = cList,
                                color="red", method.args = list(alternative = "greater"),
                                label.y = c(0.4), aes(label = ..p.signif..))+
     scale_fill_brewer(palette = "Pastel1")+
ylab("Correlation of betas")+
theme(axis.text = element_text(size=15),
              axis.title =  element_text(size=16))  
```

```r
options(repr.plot.width=8, repr.plot.height=4)

mu <- ddply(myDistFlat, "sameGene", summarise, grp.mean=mean(sDist))
options(repr.plot.width=15, repr.plot.height=7)


ggplot(myDistFlat, aes(x=sDist, color=sameGene, fill=sameGene)) +
geom_histogram(aes(y=..density..), position="identity", alpha=0.5, binwidth = 0.01)+
#geom_density(alpha=0.6)+
geom_vline(data=mu, aes(xintercept=grp.mean, color=sameGene),
           linetype="dashed")+
geom_vline(xintercept=0.015, color="black",
           linetype="dashed")+
scale_color_manual(values=c("#E69F00", "#999999",  "#56B4E9"))+
scale_fill_manual(values=c("#E69F00", "#999999" , "#56B4E9"))+
labs(title="Weight histogram plot",x="Beta correlation (pearson)", y = "Density")+
theme_bw()+theme(axis.text = element_text(size=15),
              axis.title =  element_text(size=16)) +
scale_x_continuous(breaks = seq(-0.5,0.5,0.05), labels =  seq(-0.5,0.5,0.05), limits = c(-0.5,0.5) )
```

```r
myDistFlatSameG = myDistFlat[myDistFlat$sameGene == "TRUE", ]
myDistFlatSplit <- split(myDistFlatSameG, f = myDistFlatSameG$GeneA )

myRet <- lapply(myDistFlatSplit, function(x){ if (all(x[,"sDist"] > 0.015))
                                        { 
                                           return(x)
                                     }else{
                                        x[which.max(x[,"sDist"]),]
                                   }})

myRet <- do.call(rbind, myRet)

myRet <- myRet[myRet$sDist > 0,]

myRet$guides_a <- sapply(myRet$guides_ab, function(x){kk= strsplit(x,"_")[[1]]
                                    paste0(kk[1], "_",kk[2])})

myRet$guides_b <- sapply(myRet$guides_ab, function(x){kk= strsplit(x,"_")[[1]]
                                    paste0(kk[3], "_",kk[4])})
```

```r
write.csv(myRet[,c("GeneA", "guides_a", "guides_b")], file="GuideSelect_GoodGuides.csv", row.names=FALSE, quote=FALSE)

allGuides = colnames(allWeights)
selectedGuides = unique(c(myRet$guides_a, myRet$guides_b))
badGuides <- allGuides[allGuides %ni% c(selectedGuides, "n_genes",  "mt_frac")]
badGuides <- sort(badGuides)
write.csv(badGuides, file="GuideSelect_BadKOGuides.csv", quote=FALSE, row.names=FALSE)
```
