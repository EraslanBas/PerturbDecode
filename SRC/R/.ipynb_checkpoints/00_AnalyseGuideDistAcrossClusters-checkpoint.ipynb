```r
k <- readRDS(paste0(fileDir, "/outputs/RDSFiles/GuideTablePerCluster.rds"))
k$variable <- as.character(k$variable)
```

```r
fBarMatMelted <- readRDS(paste0(fileDir, "/outputs/RDSFiles/fBarMatMelted.rds") )
```

```r
k_leiden <- as.data.frame(table(fBarMatMelted[,c("leiden")]))
```

```r
head(k_leiden)
```

```r
allTests_greater <- list()
allTests_less <- list()

depletedGuides <- list()
noOfCells = sum(k_leiden$Freq)

for (elem in as.character(unique(k$variable))){
    fBarMTemp <- k[k$variable == elem,]
    noOfGuideCells = sum(fBarMTemp$Freq)
    
    if(all(fBarMTemp$Freq == 0)){
        depletedGuides <- lappend(depletedGuides, elem)
    }else{
        
        for(cluster in 0:9){
           myDF = data.frame(clusterNo = c(k_leiden[k_leiden$Var1 == cluster,"Freq"], fBarMTemp[fBarMTemp$leiden == cluster, "Freq"]),
                  restNo = c(noOfCells - k_leiden[k_leiden$Var1 == cluster,"Freq"], noOfGuideCells - fBarMTemp[fBarMTemp$leiden == cluster, "Freq"]))
            
           testRes_g <- fisher.test(x= myDF, alternative = "greater")
           allTests_greater <- lappend(allTests_greater, c(elem, cluster,testRes_g$p.value))
            
           testRes_l <- fisher.test(x= myDF, alternative = "less")
           allTests_less <- lappend(allTests_less, c(elem, cluster,testRes_l$p.value))
        }
        
    }  
}
    
```

```r
allTests_greater <- data.frame(do.call(rbind, allTests_greater), stringsAsFactors = F)
allTests_less <- data.frame(do.call(rbind, allTests_less), stringsAsFactors = F)
```

```r
allTests_greater$FDR <- p.adjust(allTests_greater$clusterNo)
allTests_less$FDR <- p.adjust(allTests_less$clusterNo)

head(allTests_greater)
```

```r
k$KOGene <- sapply(k$variable, function(x){strsplit(x,"_")[[1]][1]})
k$KOGeneLeiden <- paste0(k$KOGene,"_",k$leiden)
```

```r
k <- data.table(k)
```

```r
k[,geneFreq := sum(Freq),by=KOGeneLeiden]
k <- data.frame(k)
```

```r
k2 <- unique(k[,c("leiden", "KOGene", "geneFreq")])
```

```r
allTests_depleted <- list()
allTests_enriched <- list()

depletedGuides <- list()
noOfCells = sum(k_leiden$Freq)

for (elem in as.character(unique(k2$KOGene))){
    fBarMTemp <- k2[k2$KOGene == elem,]
    noOfGuideCells = sum(fBarMTemp$geneFreq)
    
    if(all(fBarMTemp$geneFreq == 0)){
        depletedGuides <- lappend(depletedGuides, elem)
    }else{
        
        for(cluster in 0:9){
           
           myDF = data.frame(clusterNo = c(k_leiden[k_leiden$Var1 == cluster,"Freq"], fBarMTemp[fBarMTemp$leiden == cluster, "geneFreq"]),
                  restNo = c(noOfCells - k_leiden[k_leiden$Var1 == cluster,"Freq"], noOfGuideCells - fBarMTemp[fBarMTemp$leiden == cluster, "geneFreq"]))
            
           testRes_g <- fisher.test(x= myDF, alternative = "greater")
           allTests_depleted <- lappend(allTests_depleted, c(elem, cluster,testRes_g$p.value))
            
           testRes_l <- fisher.test(x= myDF, alternative = "less")
           allTests_enriched <- lappend(allTests_enriched, c(elem, cluster,testRes_l$p.value))
        }
        
    }  
}
  
```
